```python
import pandas as pd

df = pd.read_csv(
    "https://practicum-content.s3.us-west-1.amazonaws.com/datasets/everpeak_retail.csv"
)
```

```python
# reemplazar sentinels
df['customer_age'] = df['customer_age'].replace([-999, 999, 0, -1], pd.NA)

# imputar valores
df['customer_age'] = pd.to_numeric(df['customer_age'], errors='coerce')
df['customer_age'].fillna(df['customer_age'].median(), inplace=True)

df['city'].fillna('unknown', inplace=True)
df['state'].fillna('unknown', inplace=True)

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df = df.dropna(subset=['order_date'])
```

```python
df.to_csv("everpeak_clean.csv", index=False)
```

```python
# Funci√≥n para clasificar clientes
def classify_segment(row):
    age = row['customer_age']
    spend = row['order_value']

    # Segmentaci√≥n de Alto Valor (Gasto >= 10000)
    if spend >= 10000:
        return "VIP"
    # Segmentaci√≥n de Valor Medio (Gasto entre 5000 y 9999)
    elif spend >= 5000:
        return "Medium Value"
    # Segmentaci√≥n de Valor Bajo (Gasto < 5000)
    else:
        return "Low Value"

# aplicar funci√≥n y ver cambios
df["customer_segment"] = df.apply(classify_segment, axis=1)
df[["customer_age", "order_value", "customer_segment"]].head()
```

```python
# Funci√≥n para clasificar clientes
def classify_segment(row):
    age = row['customer_age']
    spend = row['order_value']

    # Segmentaci√≥n de Alto Valor (Gasto >= 10000)
    if spend >= 10000:
        if age >= 55:
            return "Senior VIP"
        else: # age < 55
            return "Junior VIP"

    # Segmentaci√≥n de Valor Medio (Gasto entre 5000 y 9999)
    elif spend >= 5000:
        if age >= 55:
            return "Sr. Medium Value"
        else: # age < 55
            return "Jr. Medium Value"

    # Segmentaci√≥n de Valor Bajo (Gasto < 5000)
    else: # spend < 5000
        return "Low Value"

# aplicar funci√≥n y ver cambios
df["customer_segment"] = df.apply(classify_segment, axis=1)
df[["customer_age", "order_value", "customer_segment"]].head()
```

```python
# Funci√≥n para clasificar clientes
def classify_segment(row):
    age = row['customer_age']
    spend = row['order_value']

    # Manejo de valores nulos/faltantes
    if pd.isna(age) or pd.isna(spend):
        return "Error en Datos"

    # Segmentaci√≥n
    if spend >= 10000 and age >= 55:
        return "Senior VIP"
    elif spend >= 10000 and age < 55:
        return "Junior VIP"
    elif spend >= 5000 and age >= 55:
        return "Sr. Medium Value"
    elif spend >= 5000 and age < 55:
        return "Jr. Medium Value"
    else:
        return "Low Value"
```
